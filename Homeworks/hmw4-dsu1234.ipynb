```python
#Image 1 - Transformation
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
imageOne = cv2.imread('dsu1.jpg')
pointsOne = np.array([[320,280],[760,310],[325,535],[765,505]])
imageTwo = cv2.imread('dsu1.jpg')
pointsTwo = np.array([[320,280],[760,280],[325,505],[760,505]])
h,status = cv2.findHomography(pointsOne,pointsTwo)
newImage = cv2.warpPerspective(imageOne,h,(imageTwo.shape[1],imageTwo.shape[0]))
fig, axs = subplots(1, 2, figsize=(16, 16))
axs[0].imshow(imageOne)
axs[0].axis('off')
title('Corrected Image')
axs[1].imshow(newImage)
axs[1].axis('off')
plt.show()
```
