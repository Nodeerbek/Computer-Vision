```python
#Image 5 - Transformation
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
imageOne = cv2.imread('dsu5.jpg')
pointsOne = np.array([[525,170],[700,235],[525,215],[700,270]])
imageTwo = cv2.imread('dsu5.jpg')
pointsTwo = np.array([[525,170],[700,170],[525,270],[700,270]])
h,status = cv2.findHomography(pointsOne,pointsTwo)
newImage = cv2.warpPerspective(imageOne,h,(imageTwo.shape[1],imageTwo.shape[0]))
fig, axs = subplots(1, 2, figsize=(16, 16))
rgbImage = cv2.cvtColor(newImage, cv2.COLOR_BGR2RGB)
axs[0].imshow(imageOne)
axs[0].axis('off')
title('Image corrected to read banner')
axs[1].imshow(rgbImage)
axs[1].axis('off')
plt.show()
```
